```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/pythonai250824/02.04.2025/main/sport_athletes.csv")
df
```

```python
# .1. עבור כל ענף ספורט- הצג את המשקל הנמוך ביותר , הגובה המקסימלי ומס' הזכיות המקסימליות
# a. שנה את שם עמודה weight ל- max_weight
# groups_weight = df.groupby('Sport Type').agg({
#     'Weight(kg)': 'min', 'Height(cm)': 'max', 'Wins': 'max'}).rename(columns={'Weight(kg)':'max_weight'}).reset_index()
# display(groups_weight)

# # b. שנה את שם עמודה height ל - max_height
# groups_height = df.groupby('Sport Type').agg({
#     'Weight(kg)': 'min', 'Height(cm)': 'max', 'Wins': 'max'}).rename(columns={'Height(cm)':'max_height'}).reset_index()
# display(groups_height)

# # c. שנה את שם העמודה wins ל - max_wins
# groups_wins = df.groupby('Sport Type').agg({
#     'Weight(kg)': 'min', 'Height(cm)': 'max', 'Wins': 'max'}).rename(columns={'Wins':'max_wins'}).reset_index()
# display(groups_wins)

# ככה ,a+b+c-אפשר גם את כולם יחד  
groups_Sports = df.groupby('Sport Type').agg({
    'Weight(kg)': 'min', 'Height(cm)': 'max', 'Wins': 'max'}).rename(columns={'Weight(kg)':'max_weight', 'Height(cm)':'max_height', 'Wins': 'max_wins'}).reset_index()
display(groups_Sports)
```

```python
# .2. מכל ענף ספורט - הצג את 2 המשתתפים ה גבוהים ביותר 
# מראה את 13 הראשונים הכי גבוהים, ששני הגבוהים הכי ראשונים בטבלה
display(df.sort_values('Height(cm)', ascending=False).groupby('Sport Type').head(2).reset_index(drop=True))
print()

# מראה רק את שני הראשונים הכי גבוהים
display('2 Highest Participants', df.nlargest( 2 , 'Height(cm)'))
```

```python
# .3 צור groups לפי מדינה + ענף ספורט
# a. הצג את הסטטיסטיקות הגבוהות ביותר
groups = df.groupby(['Country', 'Sport Type']).max()
display('a.statistics', groups)

# b. הצג את הסטטיסטיקות הגבוהות ביותר רק עבור כדורסל בארה"ב.רמז loc
USA_Basketball = groups.loc[('USA', 'Basketball')]
display('b.USA Basketball' ,USA_Basketball)
print()
# c. הצג את הסטטיסטיקות הגבוהות ביותר עבור כדורגל בלבד. רמז xs
Soccer = groups.xs('Soccer', level='Sport Type')
display('c.Soccer', Soccer)
print()

# d. הצג את הסטטיסטיקות הגבוהות ביותר עבור טניס וכדורגל בלבד. רמז isin
Tennis_and_Football = df[df['Sport Type'].isin(['Tennis', 'Soccer'])].groupby(['Country', 'Sport Type']).max()
display('d.Tennis_and_Football',Tennis_and_Football)
print()

# e. הפוך את האינדקסים לפי ספורט-מדינה ומיין את המידע לפי ספורט+מדינה רמז swaplevel
swaplevel = df.groupby(['Country', 'Sport Type']).max().swaplevel().sort_index(level=['Sport Type', 'Country']) 
display('e.swaplevel', swaplevel)
print()

# f. הצג את הסטטיסטיקות של ערך: מינימילי, מקסימלי, ממוצע רמז agg
statistics = groups.agg({'Height(cm)': ['min', 'max', 'mean'], 'Weight(kg)': ['min', 'max', 'mean'], 'Wins':['min', 'max', 'mean']})
display('f.minimum+maximum+mean',statistics)
```

```python
# .4. הוסף את השורות הבאות לטבלה באמצעות concat:
new_rows = pd.DataFrame({ 'Country': ['Canada', 'USA'], 'Sport Type': ['Hockey', 'Skiing'],
'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'], 'Height(cm)': [183, 172],
'Weight(kg)': [84, 68], 'Wins': [3, 4], 'Olympic Date': ['01/02/2002', '12/02/2010'] })

display(pd.concat([new_rows], axis=0))  
```

```python
# .5. שאלת merge
df_athletes_smok=pd.read_csv("https://raw.githubusercontent.com/pythonai250824/02.04.2025/main/athletes_smoking.csv")
display(df_athletes_smok)
```

```python
# .5. שאלת merge -המשך
# אחד את הטבלאות שיופיעו רק ספורטאים שיש תיעוד אם הם מעשנים או לא . רמז: inner
merge_inner = pd.merge(df, df_athletes_smok, on='Athlete Name', how='inner')
display('merge_inner', merge_inner)
print()

# אחד את הטבלאות כך שיופיעו כל הספורטאים ותיעוד אם הם מעשנים או לא, היכן שאין תיעוד -- שיופיע NaN. רמז: outer
merge_outer = pd.merge(df, df_athletes_smok, on='Athlete Name', how='outer')
display('merge_outer', merge_outer)
```
